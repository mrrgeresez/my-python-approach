## Exceptions

------

When during the execution of a program, if an unexpected situation occurs it is important to foresee what response will be provided.
The normal course to follow is to enter the `try` statement, where it is watched and checked if there is an exception in it, if it matches the `except` clause it will enter this one.

- The `except TypeError` is used to handle a specific error.
- `else` code to be executed in case there is no error.
- `finally` will always be executed.

```python
try: # Block of code to "watch".
    text = input('Type :') # enter data

except KeyboardInterrupt: # catch interrupt exception 
    print('Ctrl+c has been pressed') # Interrupt when pressing Ctrl+c

else: # executed if no error
    print('You typed {}'.format(text)) # shows entered string
 
finally: # executes whether there is an error or not
    print('end of program') # display final message
```
